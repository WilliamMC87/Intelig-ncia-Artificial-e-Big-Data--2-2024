```python
import numpy as np # biblioteca para fazer calculos matematicos
from sklearn.linear_model import LinearRegression # importa
# da biblioteca sklearn o modelo de Regressão Linear
import matplotlib.pyplot as plt # biblioteca para plotar os gráficos
import pandas as pd 
```

```python
# criando o conjunto de dados
vazao = np.array([10,20,30,5,35,40]).reshape(-1,1) # cria o conjunto de dados
nivel = np.array([50,60,70,45,80,85]) # conjunto de dados para o nivel
# Cria modelo e treina o modelo de regressão linear
modelo = LinearRegression() # armazena a regressão linear na variavel modelo
modelo.fit(vazao, nivel) # treina o modelo de regressão linear
```

```python
# Teste do modelo criado
vazao_teste = np.array([100,50,40,30,20,10,0]).reshape(-1, 1)  # vazão de água em m3/min
#vazao_teste = np.array([5, 2, 7,30,50]).reshape(-1, 1)  # vazão de água em m3/min
# Fazer previsões com base na vazão de água
nivel_previsto = modelo.predict(vazao_teste) # realiza a previsão do nivel do reservatorio

# Imprimir as previsões
for i in range(len(vazao_teste)):
    print(f"Vazão: {vazao_teste[i][0]} m3/min - Nível previsto: {nivel_previsto[i]} metros")
plt.scatter(vazao, nivel, color='blue', label='Dados de Treinamento')
plt.plot(vazao_teste,nivel_previsto, color  = 'red',label='Dados de teste')
plt.xlabel('Vazão de Água (m3/min)')
plt.ylabel('Nível do Tanque (metros)')
plt.title('Relação entre Vazão de Água e Nível do Tanque')
plt.legend() # exibe a legenda do grafico
plt.grid(True) # exibe a grade do grafico
plt.show() # plota o gráfico

```

```python
#EX2
# criando o conjunto de dados
q = np.array([100,200,300,400,500,600,700]).reshape(-1,1) # cria o conjunto de dados
temp = np.array([30,35,40,45,60,100,150]) # conjunto de dados para o nivel
# Cria modelo e treina o modelo de regressão linear
modelo = LinearRegression() # armazena a regressão linear na variavel modelo
modelo.fit(q, temp) # treina o modelo de regressão linear
```

```python
# Teste do modelo criado
q_teste = np.array([110,220,330,340,410,510]).reshape(-1, 1)  # vazão de água em m3/min
#vazao_teste = np.array([5, 2, 7,30,50]).reshape(-1, 1)  # vazão de água em m3/min
# Fazer previsões com base na vazão de água
t_prevista = modelo.predict(q_teste) # realiza a previsão do nivel do reservatorio

# Imprimir as previsões
for i in range(len(q_teste)):
    print(f"Quantidade de calor: {q_teste[i][0]} cal Temp: {t_prevista[i]} ºC")
plt.scatter(q, temp, color='blue', label='Dados de Treinamento')
plt.plot(q_teste,t_prevista, color  = 'red',label='Dados de teste')
plt.xlabel('Quantidade de calor (Cal)')
plt.ylabel('Temperatura(ºC)')
plt.title('Relação entre quantidade de calor e temperatura')
plt.legend() # exibe a legenda do grafico
plt.grid(True) # exibe a grade do grafico
plt.show() # plota o gráfico
```
