```python
import pandas as pd # importa a biblioteca
#pandas 
import matplotlib.pyplot as plt # importa a matplotlib
```

```python
dados = pd.read_csv('imigrantes_canada.csv') # importa
# a base de dados
```

```python
dados.head() # exibe as 4 primeiras linhas do dataframe
```

```python
dados.tail() # exibe as 4 ultimas linhas
```

```python
dados.info() # exibe as informações da base de dados
```

```python
# seta o pais como index
dados.set_index(['País'],inplace=True) # seta a coluna Pais como index
anos = list(map(str,range(1980,2013)))  # transforma os anos em string utilizando a
# função map 
```

```python
print(anos) # exibe os anos na saida da celula
dados.tail(5)
```

```python
brasil = dados.loc['Brasil',anos] # filtro para pegar os dados somente do Brasil
# dados.loc[coluna desejada]
```

```python
brasil
```

```python
# criando dicionario com os dados do brasil
brasil_dic= {
    'ano': brasil.index.tolist(), # pega os anos
    'imigrantes': brasil.values.tolist() # pega a quantidade de imigrantes
    }
```

```python
# criando dataframe para o Brasil
df_brasil = pd.DataFrame(brasil_dic)
```

```python
df_brasil.head() # exibe os imigrantes do dataframe brasil
```

```python
# plotando o número de imigrantes do Brasil para o Canadá
plt.plot(df_brasil['ano'],df_brasil['imigrantes']) # plota o grafico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14) # ajuste de valores para o eixo x
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores para o eixo y
plt.xlabel("Anos",fontsize=14) # titulo eixo x
plt.title('Imigração do Brasil para o Canadá',fontsize =18) # adiciona título no gráfico
plt.ylabel("Número de imigrantes",fontsize=14) # titulo eixo y
plt.show() #exibe a figura
```

```python
argentina = dados.loc['Argentina',anos]
```

```python
argentina_dic = {
    'ano': argentina.index.tolist(),
    'imigrantes': argentina.values.tolist()
    }
```

```python
df_argentina = pd.DataFrame(argentina_dic)
```

```python
# plotando o número de imigrantes do Brasil para o Canadá
plt.plot(df_argentina['ano'],df_argentina['imigrantes']) # plota o grafico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14) # ajuste de valores para o eixo x
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores para o eixo y
plt.xlabel("Anos",fontsize=14) # titulo eixo x
plt.title('Imigração da Argentina para o Canadá',fontsize =18) # adiciona título no gráfico
plt.ylabel("Número de imigrantes",fontsize=14) # titulo eixo y
plt.show() #exibe a figura
```

```python
plt.figure(figsize=(10,5)) # aumenta o tamanho da figura
plt.plot(df_brasil['ano'],df_brasil['imigrantes'],color ='red',label ='Brasil')
plt.plot(df_argentina['ano'],df_argentina['imigrantes'],color ='blue',label= 'Argentina')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize= 14)
plt.yticks([0,500,1000,1500,1500,2000,2500,3000],fontsize=14)
plt.title('Imigração do Brasil e Argentina para o Canadá',fontsize=14)
plt.ylabel("Nº de imigrantes",fontsize=18)
plt.xlabel('Anos',fontsize=14)
plt.legend() # adiciona legenda ao gráfico
plt.savefig('Dados imigração.jpg')
plt.show()
```

```python
plt.bar(df_brasil['ano'],df_brasil['imigrantes']) # plt.bar plota grafico de barras
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize= 14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize = 14)
plt.title('Imigração do Brasil para o Canadá')
plt.show()
```

```python
df_comparacao = dados.loc[['Brasil','Argentina'],anos] # filtro que pega duas colunas do 
# dataset
```

```python
df_comparacao # dataset de comparação
```

```python
df_comparacao = df_comparacao.T # aplica a matriz transposta no Dataframe
```

```python
df_comparacao['T'] = df_comparacao['Brasil'] + df_comparacao['Argentina'] # soma o total de imigrantes
```

```python
plt.figure(figsize=(10,10))
plt.plot(df_comparacao['Brasil'],label='Brasil',color='green')
plt.plot(df_comparacao['Argentina'],label='Argentina',color='blue')
plt.plot(df_comparacao['T'],label = 'Total',color = 'red')
plt.legend()
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.title('Imigração do Brazil e \n Argentina para o Canadá',fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8)) # cria a variavel ax para armazenar o subplot
ax.plot(df_brasil['ano'],df_brasil['imigrantes'],lw=3,color='green',label='Brasil') 
# ax - axes por conta que agora foi criada uma variavel
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # seta no eixo x o espaçamento do gráfico
ax.xaxis.set_tick_params(labelsize=12) # altera o tamanho da fonte do gráfico
ax.yaxis.set_tick_params(labelsize=12) # altera o tamanho da fonte do eixo y
ax.set_xlabel('Anos',fontsize=12)
ax.set_ylabel('Número de imigrantes',fontsize= 12)
ax.set_title("Imigrantes do Brasil para o Canadá",fontsize =18)
#ax.legend(loc='upper right') # posicionamento da legenda
#ax.legend(loc='upper left') # posicionamento da legenda
#ax.legend(loc='lower right') # posicionamento da legenda
#ax.legend(loc='lower left') # posicionamento da legenda
ax.legend(loc='best')
ax.grid() # adiciona o grid no gráfico
```

```python
# cria um subplot com duas figuras no gráfico
fig, axs = plt.subplots(1,2,figsize=(15,5)) # cria uma variavel axs 
# que corresponde a eixos diferentes no gráfico
# subplot (qtde de linhas, qtde de colunas, figsize - tamanho da figura)
axs[0].plot(df_brasil['ano'],df_brasil['imigrantes']) # plota o primeiro gráfico na posição 0
axs[0].set_title('Imigrantes do Brasil para o Canadá de \n 1980 - 2013',fontsize=18)
axs[0].set_xlabel('Anos',fontsize=12)
axs[0].set_ylabel('Número de imigrantes',fontsize=12)
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5)) # cria o intervalo de 5 em 5 no eixo x
axs[0].grid() # cria o grid do gráfico

# gráfico 2
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Box plot da imigração do Brasil \n para o Canadá',fontsize=18)
axs[1].set_xlabel('Brasil',fontsize=12)
axs[1].set_ylabel('Nº de imigrantes',fontsize=12)
axs[1].grid()

plt.show()
```
